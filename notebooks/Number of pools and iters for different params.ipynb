```python
from golfy import init, optimize, violations_per_replicate

results = {}
for num_peptides in range(100, 1100, 200):
    for peptides_per_pool in range(10, 50, 10):
        config = (num_peptides, peptides_per_pool)
        print("!!!", config)
        s = init(num_peptides, peptides_per_pool, strategy='greedy')
        max_iters = 1000
        _, history = optimize(s, verbose=True, return_history=True, max_iters=max_iters)
        results[config] = (len(history), s.num_pools())
```

```python
results
```

```python
import pandas as pd; 
df = pd.DataFrame({
    'num_peptides': [
        k[0] for k in results.keys()], 
    'peptides_per_pool': [
        k[1] for k in results.keys()],
    'iters': [v[0] for v in results.values()],
    'pools': [v[1] for v in results.values()],
});
df
```
